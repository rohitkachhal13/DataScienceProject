### 4. Data Preprocessing

#### a). Dealing with null values

```python
df.info()
```

##### Outcome : 
* Gender, Married, Dependents, Self_Employed, Loan_Amount_Term, Credit_History are categorical data and having null values.
* LoanAmount is continous data and having null values.

```python
# Replacing null values by median in case of continous data
df['LoanAmount'].median()
```

```python
df['LoanAmount']  = df['LoanAmount'].fillna(df['LoanAmount'].median())
```

```python
# Replacing null values by mode in case of categorical data
print('The mode of Gender is :',df['Gender'].mode()[0])
print('The mode of is Married:',df['Married'].mode()[0])
print('The mode of is Dependents:',df['Dependents'].mode()[0])
print('The mode of is Self_Employed:',df['Self_Employed'].mode()[0])
print('The mode of is Loan_Amount_Term:',df['Loan_Amount_Term'].mode()[0])
print('The mode of is Credit_History:',df['Credit_History'].mode()[0])
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Married'] = df['Married'].fillna(df['Married'].mode()[0])
df['Dependents'] = df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed'] = df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0])
df['Credit_History'] = df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
from scipy.stats import zscore
z_score = zscore(df[['ApplicantIncome','CoapplicantIncome', 'LoanAmount','Loan_Amount_Term']])
print(np.where(z_score>3))
```

### 12. Observing best random state & split the data using best random state

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
lr = RandomForestClassifier()
max_accuracy = 0
max_rs = 0
for i in range(1,100):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=i)
    lr.fit(X_train,y_train)
    lr_pred_test = lr.predict(X_test)
    accuracy = accuracy_score(y_test,lr_pred_test)
    if accuracy>max_accuracy:
        max_accuracy = accuracy
        max_rs = i
print('Maximum accuracy is :',max_accuracy,'On random state :',max_rs)
```

```python
X_c_train, X_c_test, y_c_train, y_c_test = train_test_split(X_scaled, y, test_size=0.30, random_state=max_rs)
```

```python
print('The shape of X-train is :',X_c_train.shape)
print('The shape of y-train is :',y_c_train.shape)
print('The shape of X-test is :',X_c_test.shape)
print('The shape of y-test is :',y_c_test.shape)
```

### 16. Hyperparameter Tunning on 'Random Forest Classifier' using GridSearch CV
