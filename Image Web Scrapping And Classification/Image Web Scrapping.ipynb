```python
import selenium
import warnings
import time
import urllib.request
import requests
import shutil

from selenium import webdriver
from selenium.webdriver.support import wait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
driver = webdriver.Chrome(r'E:/DATA TRAINED/Sofrware/chromedriver_win32/chromedriver.exe')



def img_save(keyword,path):
    
    driver.get("https://amazon.in")

    search_input = driver.find_element_by_xpath('//input[@id="twotabsearchtextbox"]')
    search_input.send_keys(keyword)
    search_input.send_keys(Keys.ENTER)
    time.sleep(3)
    
    
    
    image_no=1
    for i in range(0,7):
        img= driver.find_elements_by_xpath('//img[@class="s-image"]')

        for j in img:

            url= j.get_attribute("src")


            try:

                name = path + str(image_no) +"'" +'.jpg'
                res = requests.get(url, stream = True)
                if res.status_code == 200:
                    with open(name,'wb') as file:
                        shutil.copyfileobj(res.raw, file) 
                    print('Image sucessfully Downloaded: ',image_no)
                    image_no+=1


                else:
                    print('Image Couldn\'t be retrieved') 


            except:
                continue

        driver.find_element_by_xpath('//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]').click()     
        time.sleep(3)                
        
try:
    img_save("sarees",r"C://Users//Rohit Kachhal//Data Science Program//Internship//Final For Submission//Image Classification//Pictures//Sarees_image//file_'")
except:
    print("******************************************************")
try:
    img_save("Trousers",r"C://Users//Rohit Kachhal//Data Science Program//Internship//Final For Submission//Image Classification//Pictures//Trousers_image//file_'")
    
except:
    print("******************************************************")
try:
    img_save("Jeans",r"C://Users//Rohit Kachhal//Data Science Program//Internship//Final For Submission//Image Classification//Pictures//Jeans_image//file_'")
except:
    print("******************************************************")
```
